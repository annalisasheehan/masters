## MAIAC AP 

```python
AP_Monthly = pd.read_csv(r'D:\Annies_Dissertation\Analysis\Regression\Monthly_PM25_LSOA.csv')
```

```python
AP_Monthly = AP_Monthly.rename(columns={'LSOA':'lsoa11', 'value':'PM25'})
```

```python
AP_Monthly = AP_Monthly[['time', 'lsoa11', 'PM25', 'month', 'year']].dropna()
```

## Regression analysis

```python
MAIAC_HES_Soton_R = pd.merge(Soton_Resp, AP_Monthly, on=['lsoa11' , 'year', 'month'], how='inner')
```

```python
MAIAC_HES_Soton_R
```

```python
MAIAC_HES_Soton_R.corr()
```

```python
model = ols("DSR ~ PM25", data=MAIAC_HES_Soton_R)
results = model.fit()
results.summary()
```
